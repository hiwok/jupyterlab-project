```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example: Create a simple DataFrame
df = pd.DataFrame({
    'x': np.linspace(0, 10, 100),
    'y': np.sin(np.linspace(0, 10, 100))
})
df.head()
```

```python
# Example: Simple plot
plt.plot(df['x'], df['y'])
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
